```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as numpy
```

```python
mortality_csv = "detailed_mortality.csv"
df = pd.read_csv(mortality_csv)
mortality_summary=df[["Year","State", "Ten-Year Age Groups", "Race", "Population", "Deaths", "Gender"]]
mortality_summary["Year"]= mortality_summary["Year"].astype(int)
reduced_mortality = mortality_summary.dropna(how="any")
reduced_mortality.head()
```

# Deaths by Year Data

```python
#year_summary = reduced_mortality_df.set_index(["Year"])
yeardata = reduced_mortality[["Year", "State","Population", "Deaths"]]
deathperyear = yeardata.groupby("Year")["Deaths"].sum()
year_summary_df = pd.DataFrame({"Total Opioid Death by Year": deathperyear})
year_summary_df
#visualization: bar chart? Line ?
```

# Death by State, Year

```python
state_yearsummary = reduced_mortality[["Year", "State", "Population", "Deaths"]]
deathsbystateyear =state_yearsummary.groupby(["State", "Year"])["Deaths"].sum()
deathsbystateyear.unstack()
#visualization  = tableau (V)
```

# GroupBy Age

```python
age_summary = reduced_mortality[["Year", "State", "Deaths", "Ten-Year Age Groups"]]
agebyyear = age_summary.groupby(["Ten-Year Age Groups","Year"])["Deaths"].sum()
agebyyear.unstack()
#visualization = clustered bar? / multi line 
```

# GroupBy Gender

```python
gender_summary = reduced_mortality[["Year", "State", "Deaths", "Gender"]]
genderbyyear = gender_summary.groupby(["Gender", "Year"])["Deaths"].sum()
genderbyyear.unstack()
```
