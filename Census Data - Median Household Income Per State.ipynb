```python
census_09_df = pd.DataFrame(census_data09)
census_09_df["Year"] = 2009
census_10_df = pd.DataFrame(census_data10)
census_10_df["Year"] = 2010
census_11_df = pd.DataFrame(census_data11)
census_11_df["Year"] = 2011
census_12_df = pd.DataFrame(census_data12)
census_12_df["Year"] = 2012
census_13_df = pd.DataFrame(census_data13)
census_13_df["Year"] = 2013
census_14_df = pd.DataFrame(census_data14)
census_14_df["Year"] = 2014
census_15_df = pd.DataFrame(census_data15)
census_15_df["Year"] = 2015
census_16_df = pd.DataFrame(census_data16)
census_16_df["Year"] = 2016
```

```python
a = pd.merge(census_09_df,census_10_df,how="outer")
aa = pd.merge(a, census_11_df, how = "outer")
aaa = pd.merge( aa, census_12_df, how="outer")
b = pd.merge(aaa, census_13_df, how = "outer")
bb = pd.merge( b, census_14_df, how = "outer")
bbb = pd.merge( bb,census_15_df, how="outer")
census_y = pd.merge(bbb, census_16_df, how="outer")
```

```python
census_y = census_y[["NAME","B01003_001E","B17001_002E","B19013_001E","B19301_001E","Year"]]

census_y.head()
```

```python
census_df = census_y.rename(columns = {
    "NAME":"State",
    "B01003_001E":"Population",
    "B17001_002E":"Number of Persons in Poverty",
    "B19013_001E": "Median Household Income",
    "B19301_001E": "Per Capita Income"
})
census_df.head()
```

```python
census_df["Poverty Rate"] = (census_df["Number of Persons in Poverty"].astype(int) / census_df["Population"].astype(int) )*100
```
